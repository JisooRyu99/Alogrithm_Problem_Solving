```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
print(diabetes.data.shape, diabetes.target.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(diabetes.data[0:,2],diabetes.target)
plt.show()
```

```python
plt.scatter(diabetes.data[0:,2],diabetes.target)
plt.xlabel('x')
plt.ylabel('target')
plt.show()
```
